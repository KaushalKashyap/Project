```python
# Now in above genre you can see that data are in the dictionary inside list format.
# and we need only the name bcz we have id from movie id feature. 
# so for that , we are using 'ast' library which is dedicated for fetching data what you need..
import ast
```

```python
# here we are only keeping the name and leave the keywords..
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
# now here you can see genre store only name feature..
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
# same things with keywords too.
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# Now here we are getting only those who are in top 3 (top in case of indexing)
# in Cast we are looking for name , bcz every actore character name is different but real name will be the same everywhere.
# so fetching data based name will retrieve the correct result..
def convert1(text):
    L = []
    counter =0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter +=1        
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
# so here we are fetching the director data, because director name are more famous than any other entity like editor,recorder...
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] =='Director':
            L.append(i['name'])
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(20)
```

```python
# Now here is to define which data is for role  
# Movie_id, title, tags(genres	keywords , overview, 	cast	crew)
```

```python
# Now here creating function for removing the space between all the character features
# bcz same name from different movies or different last name (ex- sam worthwington, sam reign) will create redundant data,
# which effect on result and the model too..
# just for unique identity of every data..
# one another reason is that space actually tell machine that this is two different word 
# example (science fiction), but actually is a single meaning words.
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# here we are applying the function to these features ..
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
# here we are converting the whole paragraph to words by splitting it with (,)..
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
# Now here we are storing overview genre,keywords, cast and crew data into new variable ..
# so calling and fetching for recommendation will become easy...
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] +   movies['crew']
```

```python
# Now here created tags title and movie id, so, we are no longer needed of overview genre,keywords,cast, crew feature,,
final_df = movies.drop(columns = ['overview','genres', 'keywords','cast','crew'])
```

```python
# above you can see that tags feature is actually in list format []..
# romoving that is important..
final_df['tags']= final_df['tags'].apply(lambda d: "  ".join(d) )
```

```python
# In Recommendation system, there is a term called Tokenization..
# It means if there any paragraph in text format you have to convert it into sentences, then convert into words 
# and then convert into vectors..
# This is bcz with converting into vectors you won't be able to fetch and find similar result 
# or you can say it willnot be able recommend according to the input..
# another thing is remove the stopwords (this,is am, in of, for,it,there,was were kind of words)
# another thing is we have to convert into case sensitive , bcz python is case sensitive language and without converting it lower case
# it may contain redundant data like ( Audio,audio).. so convert it into lower case..
```

```python
import nltk
# nltk.download()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stopwords(obj):
    final_sentence = []
    tokens = word_tokenize(obj)
    
    for each_word in tokens:
        if each_word not in stopwords.words('english'):
            final_sentence.append(each_word)
    return " ".join(final_sentence)
```

```python
final_df['tags'] =  final_df['tags'].apply(remove_stopwords)
```

```python
final_df['tags'].head()[0]
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    result =[]
    for i in text.split():
        result.append(ps.stem(i))
    return " ".join(result)
```

```python
final_df['tags'] = final_df['tags'].apply(stemming)
```

```python
final_df['title'] = final_df['title'].str.lower()
```

```python
def find_closest(text):
    spliter = final_df['title'].str.split(' ')
    i = 0
    for val in spliter:
        if text in val:
            print(final_df['title'][i])
            return i
        i+=1
```

```python
def recommend_with_bert(movie):
    movie = movie.lower()
    titles = final_df['title']
    if movie in titles.to_list():
        index = final_df[final_df['title']==movie].index[0]
    else:
        index = find_closest(movie)
        
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x: x[1])
    for i in distances[1:10]:
        print(final_df.iloc[i[0]].title)
```
